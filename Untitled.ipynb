```python
import  pyautogui
from pynput import keyboard
from multiprocessing import Queue
from threading import Thread
import time
queue = Queue()


def clicks():
    clicking = False
    while True:
        if not queue.empty():
            value = queue.get()
            if value == "start":
                clicking = True
            elif value == "stop":
                clicking = False
        if queue.empty() and clicking:
            pyautogui.click(clicks=3)

Thread(target=clicks).start()


def on_press(key):
    if key == keyboard.KeyCode.from_char('m'):
        print("started")
        queue.put("start")
    if key == keyboard.KeyCode.from_char('n'):
        print("stopped")
        queue.put("stop")
        
listeners = keyboard.Listener(
    on_press=on_press,
    )
listeners.start()
    
        
```

```python
import os
from elasticsearch import Elasticsearch

# Initialize Elasticsearch client
es = Elasticsearch("http://localhost:9200")

def search_keys_in_documents(index):
    hits = []
    keys_list = ["ФИО", "Адрес", "Дата Рождения"]
    for key in keys_list:
        query = {
            "query": {
                "match": {
                    "content": key
                }
            }
        }
        res = es.search(index=index, body=query,  _source=True, docvalue_fields=['content'])  # Retrieving the source document
        if res['hits']['total']['value'] > 0:
            print(f"Key '{key}' found in documents.")
            for hit in res['hits']['hits']:
                hits.append(hit)
                # Creating a new document in the "leak" index
                es.index(index="file_leaks", body=hit['_source'])
                
        else:
            print(f"Key '{key}' not found in documents.")
        return hits
```

```python
import os
from elasticsearch import Elasticsearch

# Initialize Elasticsearch client
es = Elasticsearch("http://localhost:9200")

# Function to search for sensitive data in a file using Elasticsearch
def search_keys_in_documents(index):
    keys_list = ["ФИО", "Адрес", "Дата Рождения"]
    for key in keys_list:
        query = {
            "query": {
                "match": {
                    "content": key
                }
            }
        }
        res = es.search(index=index, body=query, _source=False, docvalue_fields=['content'])
        if res['hits']['total']['value'] > 0:
            print(f"Key '{key}' found in documents.")
            return res, key
        else:
            print(f"Key '{key}' not found in documents.")
    return False,False
# Call the function to search for specific keys in documents

# Function to index files in Elasticsearch
def index_files(directory, index):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.txt'):
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding="utf-8") as file_content:
                    content = file_content.read()
                    doc = {
                        'file_path': file_path,
                        'content': content
                    }
                    es.index(index=index, body=doc)



def create_index_with_mapping(index_name):
    mapping = {
        "mappings": {
            "properties": {
                "file_path": {"type": "keyword"},
                "content": {"type": "text", "fielddata": True}  # Enable fielddata for text field
            }
        }
    }
    es.indices.create(index=index_name, body=mapping)

# Main function
# if __name__ == "__main__":
#     create_index_with_mapping('new_f')
#     directory_to_search = 'C:\\Users\\Stepan\\Downloads'  # Change this to the directory you want to search
#     index_files(directory_to_search)
#     search_text_files(directory_to_search)
```
